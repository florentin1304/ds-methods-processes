# Ex1

```python
def f1(x):
    return x*np.sin(x) + 2*x

def f2(x):
    return 10*np.sin(x)+x**2

def f3(x):
    return np.sign(x)*(x**2+300) + 20*np.sin(x)
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

def generate_function(f):
    tr = 20
    n_samples = 100
    X = np.linspace(-tr, tr, n_samples)
    y = f(X)
    return X, y
    
def generate_tts(X,y):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, random_state=42, shuffle=True)
    y_test = y_test[X_test.argsort()]
    X_test.sort()
    return X_train, X_test, y_train, y_test
```

```python
fig, ax = plt.subplots(1,3)
fig.set_size_inches(15, 4)

for i,f in enumerate([f1,f2,f3]):
    X,y = generate_function(f)
    X_train, X_test, y_train, y_test = generate_tts(X, y)
    ax[i].plot(X, y)
    ax[i].scatter(X_test, y_test, color="red")

plt.show()
```
