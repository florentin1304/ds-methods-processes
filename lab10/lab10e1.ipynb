```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv("2D_gauss_clusters.csv")
# df.to_numpy()
plt.scatter(df.x, df.y)
```

```python
from KMeans import KMeans

km = KMeans(15, max_iter=50)
cl = km.fit_predict(df)
```

```python
centroids = km.centroids
plt.scatter(df.x, df.y, c=cl, cmap="jet")
plt.scatter(centroids[:, 0], centroids[:,1], color="red", marker="*")
```

# Chameleon dataset

```python
df2 = pd.read_csv("chameleon_clusters.csv")
plt.scatter(df2.x, df2.y)
```

```python
km_cham = KMeans(6, max_iter=20)
cl = km_cham.fit_predict(df2)

centroids = km_cham.centroids
plt.scatter(df2.x, df2.y, c=cl, cmap="jet")
plt.scatter(centroids[:, 0], centroids[:,1], color="red", marker="*")
```
