```python
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split, ParameterGrid, cross_val_score

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("synth.csv")
y = df.label
X = df.drop("label", axis=1)
```

```python
plt.rcParams["figure.figsize"] = (5,5)
plt.rcParams["figure.dpi"] = 200
plt.scatter(df.x0, df.x1, c=df.label)
```

# Build a model lessgo

```python
alberello = DecisionTreeClassifier()
alberello.fit(X, y)

plt.rcParams["figure.figsize"] = (20,10)
plt.rcParams["figure.dpi"] = 200
tplot = plot_tree(alberello, feature_names=X.columns)
```

The model is very deep, because it cannot divide the classifications in diagonal, so it creates "steps" that better approximate the diagonal line.
