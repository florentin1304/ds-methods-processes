```python
filename = r"C:\Users\S269491\Desktop\online_retail.csv"
df = pd.read_csv(filename)
```

```python
# filter all InvoiceNo that start with c
print(df.shape)
df_filtered = df[~df.InvoiceNo.str.contains('c')]
df_filtered = df_filtered[~df.InvoiceNo.str.contains('C')]
print(df_filtered.shape)

df_filtered = df_filtered.iloc[:20000]
```

```python
# Aggregate by invoice no in a dictionary
unique_invoices = df_filtered['InvoiceNo'].unique()
print(len(unique_invoices))
invoice_dict = {}
```

```python
df_grouped = df_filtered.groupby(by="InvoiceNo")
for x,y in df_grouped:
    invoice_dict[x] = y['Description'].tolist()
```

```python
# Get ALL the unique items and sort them
set_items = df_filtered['Description'].dropna().unique().tolist()
set_items.sort()
```

```python
# Create the matrix
pa_matrix = []
pa_indexes = []
for k,v in invoice_dict.items():
    pa_matrix.append([])
    pa_indexes.append(k)
    for item in set_items:
        pa_matrix[-1].append(True if item in v else False)
        
df_matrix = pd.DataFrame(pa_matrix, columns=set_items, index=pa_indexes)
display(df_matrix.tail())
```

```python
list_soglie = [0.5, 0.1, 0.05, 0.02, 0.01]
list_soglie = [0.05]

for i in list_soglie:
    print("="*50)
    print(i)
    fi = fpgrowth(df_matrix, i)
    print(len(fi))

fi['n'] = [len(x) for x in fi['itemsets']]
fi_sorted = fi.sort_values(by="n", ascending=False)
good_items = []

# GET ALL THE ITEMSETS THAT ARE NOT SUBSETS
for itemset in fi_sorted['itemsets'].tolist():
    add = True
    for my_itemset in good_items:
        if itemset.issubset(my_itemset):
            add = False
            break
    if add:
        good_items.append(itemset)

print(good_items)

# NON SERVE LMAO, DEVI PRENDERE SOLO QUELLO CON PIU ITEMS
interesting_itemset = fi_sorted.iloc[0]
print(interesting_itemset)
```
